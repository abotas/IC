# How to use Git for IC

```python
from IPython.display import Image
%pylab inline
```

Main repository is https://github.com/jjgomezcadenas/IC if you want to make a contribution you should fork the repository to have your own copy. There is "fork button in the upper part of the github page:

```python
Image(filename='../Images/git/fork.png') 
```

When you make a fork you will get your own git repository in your github account. For instance, mine is: https://github.com/jmbenlloch/IC

```python
Image(filename='../Images/git/forked.png') 
```

You then clone your own repository:

`git clone https://github.com/jmbenlloch/IC`

Then you can create a new branch where you will introduce the changes needed to develop the functionality you want to be included in the main repository:

`git checkout -b mybranch`

Anytime you have make a significant change that you want to be included in your github repository you can add those files and make a new commit with a brief description:

`git add example.py example2.py`

`git commit -m 'example commit'`

Next step is to upload your code to github. The first time you do it you will need to sync your new branch, which is not yet creater in your github repository:

`git push -u origin mybranch`

After that, you can do just:

`git push`

When you think your changes are ready to be included in the main repository, you need to create a new pull request:

```python
Image(filename='../Images/git/pull1.png') 
```

Check that you have chosen the correct branch. Then, after clicking on "New pull request", you will get a form where you can put a description of what you have done. You can see in the upper part that we want to merge the master branch of the main with the branch of our fork:

```python
Image(filename='../Images/git/pull2.png') 
```

The admin of the main repository will be notified and he will be able to check your changes and approve them, make some comments for you to change before accepting the pull request or close it.

When your pull request is accepted your code will be part of the main repository. If you want to continue working on another new functionality, you will probably need to update your fork to still synced with the main one.

The first time you do that you will need to include the address of the main repository:

`git remote add upstream https://github.com/jjgomezcadenas/IC`


Then you have to fetch the changes:

`git fetch upstream`

`git merge upstream/master master`

And also push them to your fork in github:

`git push`


Then you will be updated and you could start some new developments.
